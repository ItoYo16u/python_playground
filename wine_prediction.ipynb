````python
from bs4 import BeautifulSoup
import urllib3
from urllib3 import PoolManager
````

````python
PoolMG= PoolManager()
url = "http://www.liquidasset.com/winedata.html"
html = PoolMG.request("GET",url)
soup = BeautifulSoup(html.data, "html.parser")
texts=soup.text
texts=texts.split("\r\n")
del(texts[0:50]) 
texts[0]=texts[0].strip()
column_names = texts[0].split()
````

````python
splited_texts=[]
for e in texts:
    splited_texts.append(e.split())
splited_texts= splited_texts[2:-3]
````

````python
import numpy as np
vals = np.array(splited_texts)
import pandas as pd
````

````python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
````

````python

df = pd.DataFrame(vals)
df.columns = column_names
df.index = np.arange(1,39)
df = df.replace(".",np.nan)
````

````python
#df からnanを含む行を削除する
df_nan_removed=df.dropna()
df_nan_removed
````

````python
#LPRICE2を目的変数として回帰分析する
from sklearn import linear_model as LM
````

````python
LRmodel = LM.LinearRegression()
y = df_nan_removed['LPRICE2']
x = df_nan_removed[['WRAIN','DEGREES','HRAIN','TIME_SV']]
LRmodel.fit(x,y)
````
